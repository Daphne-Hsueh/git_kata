```python
# Load Titanic data and show first 10 rows

titanic = load_data()

titanic.head(10)
```

```python
# Clean the Titanic data and plot passenger class distribution
import matplotlib.pyplot as plt

# Clean the data
titanic_clean = clean_data(titanic)

# Plot passenger class distribution
plt.figure(figsize=(6,4))
titanic_clean['pclass'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Passenger Class Distribution (Cleaned Data)')
plt.show()
```
